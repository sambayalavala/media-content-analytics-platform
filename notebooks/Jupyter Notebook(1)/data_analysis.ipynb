```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import os

ROOT = r"C:\Users\samba\OneDrive\Desktop\Media_Content_Analytics_Platform"
PROC_DIR = os.path.join(ROOT, "data", "processed")

news_path = os.path.join(PROC_DIR, "NEWS_yahoo_11cols.csv")
yt_path = os.path.join(PROC_DIR, "dim_video.csv")
```

```python
news = pd.read_csv(news_path) if os.path.exists(news_path) else pd.DataFrame()
yt = pd.read_csv(yt_path) if os.path.exists(yt_path) else pd.DataFrame()

print("News shape:", news.shape)
print("YouTube shape:", yt.shape)

news.head()
```

```python
# News sentiment counts
if "sentiment_label" in news.columns:
    print(news["sentiment_label"].value_counts())
    px.pie(news, names="sentiment_label", title="News Sentiment").show()

# Top YouTube by views
if "views" in yt.columns:
    yt["views"] = pd.to_numeric(yt["views"], errors="coerce").fillna(0)
    top = yt.sort_values("views", ascending=False).head(10)
    px.bar(top, x="title", y="views", title="Top 10 YouTube by Views").show()
```
